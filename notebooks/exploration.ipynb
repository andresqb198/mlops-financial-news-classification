# Exploration Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
df = pd.read_csv('../data/train.csv')
df.head(5)
```

```python
df.columns = ['label', 'title', 'text']
df.head(5)
```

## Valores nulos y duplicados

```python
print(df.isnull().sum())
```

```python
df.duplicated().sum()
```

## Distribución de la variable target

```python
plt.figure(figsize=(8, 5))
sns.countplot(x=df['label'])
plt.title("Distribución de la Variable 'label'")
plt.xlabel("Label")
plt.ylabel("Frecuencia")
plt.show()
```

## Análisis de texto: longitud de 'title' y 'text'

```python
df["title_length"] = df["title"].astype(str).apply(len)
df["text_length"] = df["text"].astype(str).apply(len)

plt.figure(figsize=(12, 5))
sns.histplot(df["title_length"], bins=30, kde=True, color="blue", label="Title Length")
sns.histplot(df["text_length"], bins=30, kde=True, color="red", label="Text Length")
plt.legend()
plt.title("Distribución de la longitud de 'title' y 'text'")
plt.show()
```

## Nube de palabras de la variable 'title'

```python
text_title = " ".join(df["title"].dropna().astype(str))
wordcloud_title = WordCloud(width=800, height=400, background_color="black").generate(text_title)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_title, interpolation="bilinear")
plt.axis("off")
plt.title("Nube de palabras - Title")
plt.show()
```

```python
text_body = " ".join(df["text"].dropna().astype(str))
wordcloud_body = WordCloud(width=800, height=400, background_color="black").generate(text_body)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_body, interpolation="bilinear")
plt.axis("off")
plt.title("Nube de palabras - Text")
plt.show()
```
